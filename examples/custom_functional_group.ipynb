```python
from rdkit import Chem
from bonafide import AtomBondFeaturizer
```

**Objective: identify the terminal carbon atom of a terminal alkyne**

```python
smiles = "C#CCC1=CC=CC(C#CC(C)=O)=C1"

# # In case testing with explicit hydrogen atoms is desired
# mol = Chem.MolFromSmiles(smiles)
# mol = Chem.AddHs(mol)
# smiles = Chem.MolToSmiles(mol)
# smiles
```

```python
f = AtomBondFeaturizer()
f.read_input(input_value=smiles, namespace="test-alkyne")
f.show_molecule()
```

```python
# # The key level can be set to "l0", "l1" (default), or "l2". This defines how fine-grained the functional group analysis is performed.
# # However, this has no influence here, as the terminal alkyne carbon atom is not correctly identified with either key level.

# f.set_options(("bonafide.functional_group.key_level", "l1"))
```

```python
f.featurize_atoms(atom_indices="all", feature_indices=27)  # 27: functional_group_match feature
```

```python
f.return_atom_features()
```

**Problem**: both with and without hydrogens in the SMILES we cannot correctly identify the terminal alkyne carbon atom -> define custom SMARTS

```python
custom_alkyne_smarts_c = "[#6H1;$(*#[#6])]"  # This SMARTS also works for SMILES strings with explicit hydrogen atoms
custom_alkyne_smarts_h = "[#1;$(*-[#6]#[#6])]"  # This would be the analogous SMARTS for targeting the terminal alkyne hydrogen atom; works only with explicit hydrogens in the SMILES string

f.set_options(
    ("bonafide.functional_group.custom_groups", [("CustomTerminalAlkyne", custom_alkyne_smarts_c)])
)
```

```python
# Besides checking `bonafide.log`, it can be double-checked if the custom SMARTS was correctly registered.
f._functional_groups_smarts
```

```python
f.clear_atom_feature_cache()
f.featurize_atoms(atom_indices="all", feature_indices=27)
```

```python
f.return_atom_features()
```

Now, the terminal carbon atom of the terminal alkyne is correctly labeled with the custom "CustomTerminalAlkyne" label.
